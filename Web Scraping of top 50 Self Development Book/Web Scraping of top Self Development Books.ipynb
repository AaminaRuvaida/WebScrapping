# Scraping Top 50 Self Development Books.

```python
#Importing Libraries
import selenium
from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
#Opening the Chrome Browser
driver=webdriver.Chrome('chromedriver.exe')
```

```python
#Opening the URL
url="https://www.goodreads.com/list/show/152245.The_Greatest_Books_for_Personal_Growth_Self_Development_"
driver.get(url)
```

```python
#Maximize the Chrome Window
driver.maximize_window()
```

```python
#Scraping the title of the book
title=driver.find_elements_by_xpath('//a[@class="bookTitle"]')
titles=[]
for i in title:
    titles.append(i.text)
```

```python
#Scraping the book author name
author=driver.find_elements_by_xpath('//a[@class="authorName"]')
book_author=[]
for i in author:
    book_author.append(i.text)
```

```python
#Scraping the Ratings
rating=driver.find_elements_by_xpath('//span[@class="minirating"]')
ratings=[]
for i in rating:
    ratings.append(i.text.replace('it was amazing\n',"").replace('really liked it\n',""))
```

```python
r=ratings[1:]
```

```python
#Scraping the Score
score=driver.find_elements_by_xpath('//span[@class="smallText uitext"]/a')
scores=[]
for i in score:
    scores.append(i.text.replace('score:',""))
```

```python
a=scores[0:114:2]
```

```python
#Scraping the people voted
score=driver.find_elements_by_xpath('//span[@class="smallText uitext"]/a')
scores=[]
for i in score:
    scores.append(i.text.replace('people voted',"").replace('person voted',""))
```

```python
b=scores[1:114:2]
```

```python
#Scraping the Image URL
link=driver.find_elements_by_tag_name('a')
img_url=[]
for i in link:
    url=i.get_attribute('href')
    img_url.append(url)
```

```python
#Show top 50 Self Development Books
df=pd.DataFrame()
df['Book Title']=titles[:50]
df['Author Name']=book_author[:50]
df['Ratings']=r[:50]
df['Score']=a[:50]
df['People Voted']=b[:50]
df['Image URL']=img_url[:50]
df
```

```python
#Saving the data in csv file
df.to_csv('Top_50_Self_Development_Book.csv')
```
