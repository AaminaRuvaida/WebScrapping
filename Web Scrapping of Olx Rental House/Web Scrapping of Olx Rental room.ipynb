# WebScrapping of Olx

```python
#Importing the libraries
import selenium
from selenium import webdriver
import pandas as pd
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
#Opening the chrome browser
driver=webdriver.Chrome('chromedriver.exe')
driver
```

```python
#Maximize the Chrome window
driver.maximize_window()
```

```python
#Opening the URL in Chrome
url="https://www.olx.in/for-rent-houses-apartments_c1723/q-room-for-rent"
driver.get(url)
```

```python
#Scrapping the price of rental house
price=driver.find_elements_by_xpath('//span[@class="_89yzn"]')
prices=[]
for i in price:
    prices.append(i.text.replace('â‚¹',""))
```

```python
#Scrapping the description of house
description=driver.find_elements_by_xpath('//span[@class="_2tW1I"]')
desc=[]
for i in description:
    desc.append(i.text)
```

```python
#Scrapping the address of rental house
address=driver.find_elements_by_xpath('//span[@class="tjgMj"]')
add=[]
for i in address:
    add.append(i.text)
```

```python
#Scrapping the room available
BHK=driver.find_elements_by_xpath('//span[@class="_2TVI3"]')
bhk=[]
for i in BHK:
    bhk.append(i.text)
```

```python
#Scrapping the date when it is posted on the website
posted=driver.find_elements_by_xpath('//span[@class="zLvFQ"]')
pos=[]
for i in posted:
    pos.append(i.text)
```

```python
#Scrapping the Image URL
link=driver.find_elements_by_tag_name('a')
url=[]
for i in link:
    urls=i.get_attribute('href')
    url.append(urls)
```

```python
#Storing the entire data in data frame
df=pd.DataFrame()
df['Description']=desc
df['BHK']=bhk
df['Address']=add
df['Posted']=pos
df['Image URL']=url[:40]
df['Price']=prices
df
```

```python
df.to_csv('Olx_rental_house.csv')
```
